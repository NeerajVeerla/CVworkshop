```python
from sklearn.datasets import load_iris
from numpy.random import randint
import numpy
import matplotlib.pyplot as plt
numpy.random.seed(0)

#loading data
iris=load_iris()

#store feature  matrix in X
X=iris.data
#target in y
y=iris.target

#
train_idx=xxx=randint(0,150,100)
X_train=X[train_idx]
y_train=y[train_idx]

test_idx=numpy.setdiff1d(range(150),train_idx)
X_test=X[test_idx]
y_test=y[test_idx]

print(X_test)




```
